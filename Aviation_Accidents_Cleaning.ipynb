### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. **Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

Missing values in injury-related columns were carefully handled to avoid introducing bias.
Fatal, serious, and minor injuries were assumed to be zero when missing, as these are typically reported when present.
However, missing values in the total_uninjured column were not imputed, as absence of data does not imply zero uninjured passengers.
Total passengers were computed as the sum of all injury categories.
Rows with incomplete or zero passenger counts were removed to ensure valid rate calculations.

```python
#estimating the total number of passangers
df['total_passengers'] = df[[
    'total_fatal_injuries',
    'total_serious_injuries',
    'total_minor_injuries',
    'total_uninjured'
]].sum(axis=1, skipna=False)
df = df[df['total_passengers'].notna()]
df = df[df['total_passengers'] > 0]
df['total_passengers'].sum()
```

the total number of passagers is 181591
we know that the total number of passegers represent everyone present in the plain both injured and uninjured

2. The likelihood of serious/fatal injury

```python
#creating a key safety metrics
df['fatal_injury_rate'] = df['total_fatal_injuries'] / df['total_passengers']#fatal injury rate
df['serious_injury_rate'] = df['total_serious_injuries'] / df['total_passengers']#serious injury rate
df['severe_injury_rate'] = (
    df['total_fatal_injuries'] + df['total_serious_injuries']
) / df['total_passengers']#combined risk
```
